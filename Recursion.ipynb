```python
# Remove the minimum number of invalid parentheses in order to make the 
# input string valid. Return all possible results.
# https://leetcode.com/explore/interview/card/facebook/53/recursion-3/324/

def removeInvalidParentheses(s):
    output = list()
    done = list() # This is to help with the speed. Checks if string has already been checked
    output = list(set(mainParentheses(s, output, done)))
    res = list()
    for ele in output:
        if len(ele) == len(max(output, key = len)):
            res.append(ele)
    if res == []:
        s2 = list(s)
        s2 = [ele for ele in s2 if ele != '(' and ele != ')']
        res.append(''.join(s2))
    return res
    
def mainParentheses(s, output, done):
    if s not in done:
        done.append(s)
        #print(s, output)
        leng = len(list(s))
        if output != []:
            maxEle = len(max(output, key = len))
        else:
            maxEle = 0
        if check(s) == True:
                output.append(s) 
        elif leng>2 and leng-1>=maxEle: # this is so it does not keep going down
            for i in range(leng):
                s2 = list(s)
                if s2[i]=='(' or s2[i]==')': 
                    del s2[i]
                    output = mainParentheses(''.join(s2), output, done)
    return output
    
def check(s):
    '''
    Returns False if string has invalid parenthesis, True otherwise
    '''
    tmp = list()
    for c in list(s):
        if c == "(":
            tmp.append(c)
        if c == ")":
            if len(tmp)>0:
                tmp.pop()
            else:
                return(False)
    if len(tmp) > 0: 
        return(False)
    else:
        return(True)
    
def stripped(s):
    tmp = list(s)
    return ''.join([ele for ele in tmp if ele =='(' or ele ==')'])
    
#removeInvalidParentheses( "(((k()((" )
#removeInvalidParentheses( "()(((((((()" )
removeInvalidParentheses( ")(((()(y((u()(z()()" ) 
```

```python
# Given two integer arrays preorder and inorder where preorder is the preorder traversal of a binary 
# tree and inorder is the inorder traversal of the same tree, construct and return the binary tree.
# https://leetcode.com/explore/challenge/card/june-leetcoding-challenge-2021/604/week-2-june-8th-june-14th/3772/

# Solution using recursion while transversing the nodes following the preorder array

# Definition for a binary tree node.
class TreeNode:
     def __init__(self, val=0, left=None, right=None):
         self.val = val
         self.left = left
         self.right = right
class Solution:
    def buildTree(self, preorder, inorder):
        if inorder:
            tree = TreeNode(preorder[0])        # Assign head of node 
            preorder.pop(0)                      # Remove head of node from preorder
            #print(tree.val)
            pos_inorder = inorder.index(tree.val)  # since they have unique values
            left = inorder[:pos_inorder]         # inorder left tree
            right = inorder[pos_inorder+1:]      # inorder right tree. 
                                             # Skip +1 since +0 that corresponds to the head of the node
            tree.left = self.buildTree(preorder, left)
            tree.right = self.buildTree(preorder, right)
            return tree
        else:
            return None

preorder = [3,9,20,15,7]
inorder = [9,3,15,20,7]
tree = Solution().buildTree(preorder, inorder)
```
